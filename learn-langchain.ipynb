```python
from langchain.agents import create_agent
```

```python

def get_weather(lat: float, lon: float) -> str:
    """Get current weather for coordinates"""
    import requests

    url = "https://api.open-meteo.com/v1/forecast"
    params = {
        "latitude": lat,
        "longitude": lon,
        "current_weather": True
    }
    
    res = requests.get(url, params=params)
    data = res.json()
    
    return str(data["current_weather"])
```

```python
def get_coordinates(city: str) -> str:
    """Get latitude and longitude for a city"""
    import requests

    url = "https://nominatim.openstreetmap.org/search"
    params = {
        "q": city,
        "format": "json",
        "limit": 1
    }
    
    headers = {
        "User-Agent": "langgraph-app"
    }
    
    res = requests.get(url, params=params, headers=headers)
    data = res.json()
    
    if data:
        lat = data[0]["lat"]
        lon = data[0]["lon"]
        return f"{lat},{lon}"
    
    return "Not found"
```

```python
def get_joke() -> str :
    """Get a random joke"""
    import requests
    res = requests.get("https://official-joke-api.appspot.com/random_joke")
    data = res.json()
    return f"{data['setup']} - {data['punchline']}"
```

```python

from dotenv import load_dotenv
load_dotenv()  # loads GROQ_API_KEY from a .env file

from langchain_groq import ChatGroq
```

```python
model = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)

agent = create_agent(
    model=model,
    tools=[get_coordinates,get_weather, get_joke],
    system_prompt="You are a helpful assistant",
)
```

```python
result = agent.invoke(
    {"messages": [{"role": "user", "content": "Tell me a Joke and also What's the weather in Jodhpur?"}]}
)
print(result["messages"][-1])
```
